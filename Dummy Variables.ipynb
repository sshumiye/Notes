```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
# https://en.wikipedia.org/wiki/Dummy_variable_(statistics)
```

```python
train_data = pd.read_csv('kaggle_train.csv')
train_data.head()
```

```python
train_data['sex_male_var'] = train_data.Sex.map({'female' : 0, 'male':1})
train_data.head()
```

```python
dummy_data = pd.get_dummies(train_data.Sex)

dummy_data.head()
```

```python
drop_dummy = pd.get_dummies(train_data.Sex).iloc[:, 1:]
drop_dummy.head()
```

```python
pd.get_dummies(train_data.Sex, prefix = 'Sex').iloc[:, 1:].head()
```
